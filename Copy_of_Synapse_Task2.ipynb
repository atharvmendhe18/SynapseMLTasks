```python
# this was way better than the code above, cuz in the above code i would have repeated the same for loop two times
dict_movies_tvshow_per_country = {}
for i in range(5332):
  if ',' in array_after_cleaning[i,5]:
    countries_in_i = array_after_cleaning[i,5].split(',')
    for country in countries_in_i:
      country = country.strip()
      if country in dict_movies_tvshow_per_country:
        dict_movies_tvshow_per_country[country] += 1
      else:
        dict_movies_tvshow_per_country[country] = 1
  else:
    country_in_i = array_after_cleaning[i,5]
    if country_in_i in dict_movies_tvshow_per_country:
      dict_movies_tvshow_per_country[country_in_i] += 1
    else:
      dict_movies_tvshow_per_country[country_in_i] = 1
dict_movies_tvshow_per_country
```

Display a new dataframe, that contains only those movies that are present in the Top 5 countries.

(Don't forget to reset the indices again)

```python
# The length of new dataframe will be 3102, if correct
keys = list(dict_movies_tvshow_per_country.keys())
keys.sort(key = lambda x: dict_movies_tvshow_per_country[x], reverse=True)
# sorting the dict according to values
sorted_dict = {}
for key in keys:
  sorted_dict[key] = dict_movies_tvshow_per_country[key]

top_5_countries = keys[:5]
```

```python
items_to_remove_new = set({})
for i in range(5332):
  if "," in array_after_cleaning[i,5]:
    items_to_remove_new.add(i)
  elif array_after_cleaning[i,5] not in top_5_countries:
    items_to_remove_new.add(i)


top_5_movies_country = np.delete(array_after_cleaning, list(items_to_remove_new),0)
top_5_movies_country_df = pd.DataFrame(top_5_movies_country)
top_5_movies_country_df
```

## Task 2.3 - Chart

Use the new dataframe to create a Pie Chart on the basis of number of Movies/TV Shows per country.

```python
data = []
label = []
for key in keys[:5]:
  label.append(key)
  data.append(sorted_dict[key])

fig = plt.figure(figsize =(10, 7))
plt.pie(data, labels = label)
plt.show()
```
